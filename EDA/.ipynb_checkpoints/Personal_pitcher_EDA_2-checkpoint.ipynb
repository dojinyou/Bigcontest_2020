```python
import pandas as pd
import chart_studio.plotly as py
import cufflinks as cf
cf.go_offline(connected=True)
```

```python
def load_Data(name, path = "../Data/sheet_season_csv/", ):
    years = ['2016','2017','2018','2019','2020']
    name_salt = '2020빅콘테스트_스포츠투아이_제공데이터_'
    PATH = path
    df = pd.DataFrame()
    num_rows = 0
    for year in years:
        if year == '2016':
            df = pd.read_csv(PATH + name_salt+name+'_'+year+'.csv', encoding='CP949')
#             num_rows += len(df)
#             print(num_rows)
        else :
            new_df = pd.read_csv(PATH + name_salt+name+'_'+year+'.csv', encoding='CP949')
#             num_rows += len(new_df)
#             print(num_rows)
            df = pd.concat([df, new_df])
    return df
```
